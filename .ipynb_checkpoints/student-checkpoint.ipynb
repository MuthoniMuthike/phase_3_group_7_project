- **Defininig features and target variables**
- **OneHot Encoding**

```python
# Defining the features (X) and target variable (y)
X = df.drop(['churn', 'phone number', 'state'], axis=1)
y = df['churn']

# Performing one-hot encoding on categorical features
X_encoded = pd.get_dummies(X)
X_encoded.head()
```

 **Performing a train test split**

```python
# Performing train-test split with stratification to maintain the class distribution in the train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, stratify=y, random_state=42)

# Displaying the shapes of the train and test sets
print("Train set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

**Feature Scaling**

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Baseline Model

```python
# Instantiating and fitting a LogisticRegression with random_state=42

baseline_model = LogisticRegression(random_state=42)
baseline_model.fit(X_train_scaled, y_train)
```

```python
# Plotting a confunsion matrix
plot_confusion_matrix(baseline_model, X_train_scaled, y_train)
```

```python
# Using cross validation to evaluate the fiited model
cv_scores = cross_val_score(baseline_model, X_train_scaled, y_train).mean()
cv_scores
```

A cross-validation score of 0.8619 means that, on average, the model predicts the correct churn status for approximately 86.2% of the instances in the training data. This suggests that the model performs relatively well in terms of overall accuracy.

```python
# Baseline Model Metrics
y_pred = baseline_model.predict(X_test_scaled)

baseline_accuracy = accuracy_score(y_test, y_pred)
baseline_precision = precision_score(y_test, y_pred)
baseline_recall = recall_score(y_test, y_pred)
baseline_f1 = f1_score(y_test, y_pred)

print("Baseline_Accuracy:", baseline_accuracy)
print("Baseline_Precision:", baseline_precision)
print("Baseline_Recall:", baseline_recall)
print("Baseline_F1_score:", baseline_f1)
```

- An accuracy of 85.91% suggests 85.91% of the predictions made by the model on the test set are correct
- A precision of 53.49 suggests %53.49% of the customers predicted as churned by the baseline model are actually churned customers.
- A recall of 23.71% means that approximately 23.71% of the actual churned customers are correctly identified by the baseline model as churned
- The F1-score being a harmonic mean of precision and recall, considers both false positives and false negatives. In this case, the F1-score of 32.86% indicates the baseline model's performance is moderate, taking into account both precision and recall.

```python
# Baseline classification report
baseline_report = classification_report(y_test, y_pred)
# Print the metrics report
print(baseline_report)
```
